```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def semnal(frc, x):
    return np.sin(2 * np.pi * frc * x)

time_horizont = 0.02

cont_time = np.linspace(0, time_horizont, 10**5)

frc = 40

plt.plot(cont_time, semnal(frc, cont_time))

f_s = 100
disc_time = np.linspace(0, time_horizont, int(100 * time_horizont + 1))
plt.stem(disc_time, semnal(frc, disc_time))

for k in [1, 2]:
    plt.plot(cont_time, semnal(frc + k * f_s, cont_time))

plt.plot()
```

```python
def sample_with_freq(freq):
    return np.sin(2 * np.pi * 40 * np.linspace(0, time_horizont, int(time_horizont * freq) + 1))
```

```python
# plt.plot(cont_time, semnal)
plt.stem(np.linspace(0, time_horizont, int(time_horizont * 80) + 1), sample_with_freq(81))
plt.plot()
```
