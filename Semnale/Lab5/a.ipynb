```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sine(frec, t, arg):
    return np.sin(2 * np.pi * frec * t + arg)
```

```python
t = np.linspace(0, 1, 1000)
```

```python
wave1 = 2 * sine(5, t, 0)
wave2 = 5 * sine(10, t, np.pi / 4)
wave3 = 0.5 * sine(100, t, 0)

wave = wave1 + wave2 + wave3

plt.plot(wave)
```

```python
fft = np.fft.fft(wave3)
fft = np.abs(fft)
fft = fft[:len(fft) // 2]

plt.scatter([i for i in range(len(fft))], fft)
```
